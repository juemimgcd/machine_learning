```python
from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split


cancer = load_breast_cancer()

X = cancer['data']
y = cancer['target']
# 数据标准化
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 数据集划分
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.3, random_state=42
)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)
y_pred = rf.predict(X_test)

rf_accuracy = accuracy_score(y_test, y_pred)
rf_precision = precision_score(y_test, y_pred)
rf_recall = recall_score(y_test, y_pred)
rf_f1 = f1_score(y_test, y_pred)
print("随机森林模型性能:")
print(f"准确率(Accuracy): {rf_accuracy:.4f}")
print(f"精确率(Precision): {rf_precision:.4f}")
print(f"召回率(Recall): {rf_recall:.4f}")
print(f"F1分数: {rf_f1:.4f}")
```

```python
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier

bagging = BaggingClassifier(
    estimator=DecisionTreeClassifier(),
    n_estimators=100,
    random_state=42
)
bagging.fit(X_train, y_train)
y_pred = bagging.predict(X_test)

# 模型评估
bg_accuracy = accuracy_score(y_test, y_pred)
bg_precision = precision_score(y_test, y_pred)
bg_recall = recall_score(y_test, y_pred)
bg_f1 = f1_score(y_test, y_pred)

print("\nBagging模型性能:")
print(f"准确率(Accuracy): {bg_accuracy:.4f}")
print(f"精确率(Precision): {bg_precision:.4f}")
print(f"召回率(Recall): {bg_recall:.4f}")
print(f"F1分数: {bg_f1:.4f}")
```

```python
from sklearn.tree import DecisionTreeClassifier

cart = DecisionTreeClassifier(random_state=42)
cart.fit(X_train, y_train)
y_pred = cart.predict(X_test)

# 模型评估
cart_accuracy = accuracy_score(y_test, y_pred)
cart_precision = precision_score(y_test, y_pred)
cart_recall = recall_score(y_test, y_pred)
cart_f1 = f1_score(y_test, y_pred)

print("\nCART决策树模型性能:")
print(f"准确率(Accuracy): {cart_accuracy:.4f}")
print(f"精确率(Precision): {cart_precision:.4f}")
print(f"召回率(Recall): {cart_recall:.4f}")
print(f"F1分数: {cart_f1:.4f}")
```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression(random_state=42, max_iter=1000)
lr.fit(X_train, y_train)
y_pred = lr.predict(X_test)

# 模型评估
lr_accuracy = accuracy_score(y_test, y_pred)
lr_precision = precision_score(y_test, y_pred)
lr_recall = recall_score(y_test, y_pred)
lr_f1 = f1_score(y_test, y_pred)

print("\n逻辑回归模型性能:")
print(f"准确率(Accuracy): {lr_accuracy:.4f}")
print(f"精确率(Precision): {lr_precision:.4f}")
print(f"召回率(Recall): {lr_recall:.4f}")
print(f"F1分数: {lr_f1:.4f}")
```

```python
from sklearn.naive_bayes import GaussianNB

nb = GaussianNB()
nb.fit(X_train, y_train)
y_pred = nb.predict(X_test)

# 模型评估
nb_accuracy = accuracy_score(y_test, y_pred)
nb_precision = precision_score(y_test, y_pred)
nb_recall = recall_score(y_test, y_pred)
nb_f1 = f1_score(y_test, y_pred)

print("\n朴素贝叶斯模型性能:")
print(f"准确率(Accuracy): {nb_accuracy:.4f}")
print(f"精确率(Precision): {nb_precision:.4f}")
print(f"召回率(Recall): {nb_recall:.4f}")
print(f"F1分数: {nb_f1:.4f}")
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)

# 模型评估
knn_accuracy = accuracy_score(y_test, y_pred)
knn_precision = precision_score(y_test, y_pred)
knn_recall = recall_score(y_test, y_pred)
knn_f1 = f1_score(y_test, y_pred)

print("\nKNN模型性能:")
print(f"准确率(Accuracy): {knn_accuracy:.4f}")
print(f"精确率(Precision): {knn_precision:.4f}")
print(f"召回率(Recall): {knn_recall:.4f}")
print(f"F1分数: {knn_f1:.4f}")
```
